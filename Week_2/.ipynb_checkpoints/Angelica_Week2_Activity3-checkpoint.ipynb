```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
customer_df = pd.read_csv('Data_Marketing_Customer_Analysis_Round3.csv')
customer_df
```

```python
customer_df.info()
```

### Get the numeric data into dataframe called numerical and categorical columns in a dataframe called categoricals. (You can use np.number and np.object to select the numerical data types and categorical data types respectively)

```python
numericals = customer_df.select_dtypes(include='number')
numericals.info()
```

```python
categoricals = customer_df.select_dtypes(include='object')
categoricals.info()
```

### Now we will try to check the normality of the numerical variables visually
Use seaborn library to construct distribution plots for the numerical variables

```python
fig, axes = plt.subplots(3,3,figsize=(20,10))
sns.histplot(numericals.customer_lifetime_value, ax=axes [0,0])
sns.histplot(numericals.income, ax=axes [0,1])
sns.histplot(numericals.monthly_premium_auto, ax=axes [0,2])
sns.histplot(numericals.months_since_last_claim, ax=axes [1,0])
sns.histplot(numericals.months_since_policy_inception, ax=axes [1,1])
sns.histplot(numericals.number_of_open_complaints, ax=axes [1,2])
sns.histplot(numericals.number_of_policies, ax=axes [2,0])
sns.histplot(numericals.total_claim_amount, ax=axes [2,1])
```

```python
#sea plot for the categorical variables
fig, axes = plt.subplots(5,3,figsize=(20,20))
sns.countplot(categoricals.region, ax = axes[0,0])
sns.countplot(categoricals.response, ax = axes[0,1])
sns.countplot(categoricals.coverage, ax = axes[0,2])
sns.countplot(categoricals.education, ax = axes[1,0])
#sns.histplot(categoricals.effective_to_date, ax = axes[1,1])
sns.countplot(categoricals.employment_status, ax = axes[1,2])
#plt.xticks(ticks=[1,2],labels=["employed","unemployed","medical leave","disabled","retired"],rotation='45')
sns.countplot(categoricals.gender, ax = axes[2,0])
sns.countplot(categoricals.location_code, ax = axes[2,1])
sns.countplot(categoricals.marital_status, ax = axes[2,2])
sns.countplot(categoricals.policy_type, ax = axes[3,0])
#sns.histplot(categoricals.policy, ax = axes[3,1])
sns.countplot(categoricals.renew_offer_type, ax = axes[3,2])
sns.countplot(categoricals.sales_channel, ax = axes[4,0])
sns.countplot(categoricals.vehicle_class, ax = axes[4,1])
sns.countplot(categoricals.vehicle_size, ax = axes[4,2])
```

### Do the distributions for different numerical variables look like a normal distribution?

Any of the numerical variables has a normal distribution. There is no bell shape in the distribution of any of them.

### For the numerical variables, check for correlation between the input features. Note: this does not include the target feature.

```python
sns.scatterplot(x=numericals.income,y=numericals.monthly_premium_auto)
```

```python
sns.scatterplot(x=numericals.months_since_last_claim,y=numericals.number_of_open_complaints)
```

```python
sns.scatterplot(x=numericals.income,y=numericals.number_of_policies)
```

### Plot the Correlation Heatmap.

```python
fig2=plt.figure(figsize=(8,5))
sns.heatmap(numericals.corr(),annot=True)
```
