```python
#creating a function to put all headers in lower case
def lower_case_column_names(file1_df): 
    file1_df.columns=[i.lower() for i in file1_df.columns]
    return file1_df
file1_df=lower_case_column_names(file1_df)
file1_df

#another way to do it using the rename method and a lambda function
## df1 = df1.rename(columns= lambda x: x.lower())
```

```python
#Applying fuction lower_case_column_names to file 2 and checking how file 2 looks like again with update
file2_df=lower_case_column_names(file2_df)
file2_df
```

```python
#Applying fuction lower_case_column_names to file 2 and checking how file 2 looks like again with update
file3_df=lower_case_column_names(file3_df)
file3_df
```

```python
# creating a function for homologating headers and changing ' ' for '_' and checking the changes in file
def rename_columns(file1_df): 
    file1_df.rename(columns={'st':'state','customer lifetime value':'customer_lifetime_value',
                          'monthly premium auto':'monthly_premium_auto',"number of open complaints":"number_of_open_complaints", 
                          "policy type":"policy_type","total claim amount": "total_claim_amount",
                          "vehicle class":"vehicle_class"}, inplace=True )
    return file1_df
file1_df=rename_columns(file1_df)
file1_df
```

```python
# Applying function of rename_columns and applying to file 2
file2_df=rename_columns(file2_df)
file2_df
```

```python
# Applying function of rename_columns and applying to file 3
file3_df=rename_columns(file3_df)
file3_df
```

```python
#concatenating the 3 files into one
file_total_df = pd.concat([file1_df,file2_df,file3_df], axis=0)
file_total_df
```

```python
#checking the info and type of variables of the concatenated df
file_total_df.info()
```

```python
print(file_total_df['customer_lifetime_value'].isna().sum())
print(file_total_df['gender'].isna().sum())
# there are NaN values in gender and in customer_lifetime_value
# customer_lifetime_value is object type --> needs to be fixed integers or floats
# number_of_open_complaints is object type --> needs to be fixed to integers
```

```python
# Eliminating duplicates in the total file
file_total_df=file_total_df.drop_duplicates() 
file_total_df
```

```python
# resetting index after concatenation of 3 files and dropping duplicates
file_total_df = file_total_df.reset_index() 
file_total_df
```

```python
# creating a function that drops columns index (index of original 3 files) and customer
def drop_columns(file_total_df) :            
    file_total_df.drop(columns=["index","customer"], inplace=True)
    return file_total_df
drop_columns(file_total_df)
file_total_df
```

```python
#checking the total of NaN after dropping duplicates
file_total_df['customer_lifetime_value'].isna().sum()
```

```python
#checking how many times there is '%' in the column 'customer_lifetime_value'
count_percent = file_total_df['customer_lifetime_value'].str.contains('%').sum()
count_percent
```

```python
# removing % whiting the customer_lifetime_value
# Define a function to remove '%' from string values
def remove_percent(x):
    if isinstance(x, str):
        return x.strip('%')
    else:
        return x

# Apply the function to the 'customer_lifetime_value' column using .apply()
file_total_df['customer_lifetime_value'] = file_total_df['customer_lifetime_value'].apply(remove_percent)
```

```python
# counting the '%' at the customer_lifetime_value after running the function
count_percent_update = file_total_df['customer_lifetime_value'].str.contains('%').sum()
count_percent_update
```

```python
# extracting the middle number in the number_of_open_compalints that has the format of x/x/xx
## lambda function lambda x: if ... else x
file_total_df['number_of_open_complaints'] = \
file_total_df['number_of_open_complaints'].apply(lambda x: int(x.split('/')[1]) if '/' in str(x) else x)
file_total_df
```

```python
#checking the number NaNs in the customer_lifetime_value column after the merge
file_total_df['customer_lifetime_value'].isna().sum()
```

```python
#checking the index NaNs in the customer_lifetime_value column after the merge
file_total_df[file_total_df['customer_lifetime_value'].isnull()].index
```

```python
#checking the number NaNs in the number_of_open_complaints column after the merge
file_total_df[file_total_df['number_of_open_complaints'].isnull()].index
```

```python
#checking the index NaNs in the number_of_open_complaints column after the merge
file_total_df['number_of_open_complaints'].isna().sum()
```

```python
# Defining a function that will change the NaN for 0 in order to convert the column into integers in next steps.
def change_NaN(x):
    if pd.isna(x):
        return 0
    else:
        return x
# applying the function to the customer_lifetime_value
file_total_df['customer_lifetime_value'] = file_total_df['customer_lifetime_value'].apply(change_NaN)
file_total_df['customer_lifetime_value'].isna().sum()
```

```python
# applying the function to the number_of_open_complaints
file_total_df['number_of_open_complaints'] = file_total_df['number_of_open_complaints'].apply(change_NaN)
file_total_df['number_of_open_complaints'].isna().sum()
```

```python
# changing the data type to integers and checking the conversion
file_total_df['number_of_open_complaints'] = file_total_df['number_of_open_complaints'].astype(int)
file_total_df['number_of_open_complaints'].info()
```

```python
# changing the data type to integers and checking the conversion
file_total_df['customer_lifetime_value'] = pd.to_numeric(file_total_df['customer_lifetime_value'], errors='coerce')
file_total_df['customer_lifetime_value'] = file_total_df['customer_lifetime_value'].fillna(0).astype(int)
file_total_df['customer_lifetime_value'].info()
```

```python
# checking how the data type look like after changes in customer_lifetime_value and number_of_open_complaints
file_total_df
```

```python
#Filtering data and Correcting typos – Filter the data in state and gender column to standardize the \
#texts in those columns
file_total_df['gender'].unique()
```

```python
#Homogenizing the 'gender' column but leaving the NaN
gender_updated = {'F':'female','M':'male','Femal':'female','Male':'male','female':'female'}    
file_total_df['gender'] = file_total_df['gender'].replace(gender_updated)
```

```python
#checking the new unique values of column 'gender'
file_total_df['gender'].unique()
```

```python
file_total_df['state'].unique()
```

```python
#Homogenizing the 'state' column but leaving the NaN
state_updated = {'WA':'Washington','Cali':'California','AZ':'Arizona'}    
file_total_df['state'] = file_total_df['state'].replace(state_updated)
```
